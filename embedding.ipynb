```python
from sentence_transformers import SentenceTransformer

# Use a legal domain pretrained model
model = SentenceTransformer("nlpaueb/legal-bert-base-uncased")

# Example legal sentences
sentences = [
"The agreement shall terminate upon thirty (30) days written notice.",
"This contract may end if a 30-day prior notice is given.",
"The defendant filed a motion to dismiss under Rule 12(b)(6).",
"The river overflowed its banks in 1998." # irrelevant (negative example)
]

# Get sentence embeddings
embeddings = model.encode(sentences, convert_to_tensor=True)

print("Embedding shape:", embeddings.shape)
```
